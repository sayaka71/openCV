```python
# -*- coding: utf-8 -*-
"""
Created on Tue Mar 17 17:38:06 2020

@author: sayaka71
"""

# scriptã®æ›´æ–°ã‚’åæ˜ ã™ã‚‹ãŠã¾ã˜ãªã„
%load_ext autoreload
%autoreload 2

import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline


# ã‚°ãƒ©ãƒ•æç”»ã®é–¢æ•°å®šç¾©
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)

# ã‚°ãƒ¬ãƒ¼ç”»åƒ
img = cv2.imread('IMG_4453.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)


# å¤§æ´¥æ³•ã§2å€¤åŒ– retã¯boolå€¤
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

# ãƒŽã‚¤ã‚ºã‚’æ¶ˆã™
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 5)

# ç¢ºå®ŸãªèƒŒæ™¯
sure_bg = cv2.dilate(opening,kernel,iterations=5)

# --- è·é›¢å¤‰æ›ã€€ï¼ˆã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã¨èƒŒæ™¯ã®è·é›¢ã‚’å‡ºã™ï¼‰ ---
# é–¾å€¤ï¼ˆè·é›¢å¤‰æ›ã§å¾—ã‚‰ã‚ŒãŸå€¤ã®æœ€å¤§å€¤Ã—0.2ï¼‰ â†ã“ã“ã‚’èª¿ç¯€ï¼
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)
ret, sure_fg = cv2.threshold(dist_transform, 0.3*dist_transform.max(), 255, 0)

# è·é›¢å¤‰æ›ã®çµæžœã‹ã‚‰ç¢ºå®Ÿãªå‰æ™¯ã‚’
sure_fg = np.uint8(sure_fg)

# å‰æ™¯ã§ã‚‚å¾Œæ™¯ã§ã‚‚ãªã„Unknown
unknown = cv2.subtract(sure_bg,sure_fg)


# ã‚ªãƒ–ã‚¸ã‚§ã‚¯ãƒˆã”ã¨ã«ãƒ©ãƒ™ãƒ«ã‚’ã¤ã‘ã‚‹
ret, markers = cv2.connectedComponents(sure_fg)
markers = markers+1
markers[unknown==255] = 0

# markersãƒ‡ãƒ¼ã‚¿ã‚’ã‚‚ã¨ã«watershed
markers = cv2.watershed(img, markers)

# contours(è¼ªéƒ­), hierarchy(éšŽå±¤ï¼Œè¦ªå­)
contours, hierarchy = cv2.findContours(markers.copy(),cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    
    if hierarchy[0][i][3] == -1:
        # å¢ƒç•Œã‚’èµ¤è‰²ã§æã  (255, 0,ã€€0)ï¼š color red,  5pt line
        cv2.drawContours(img, contours, i, (255, 0, 0), 5)
        
display(img)
```

```python
# å¤–ã‚Œå€¤ã‚’deleteï¼ˆã‚ã¨ã§å¤–ã‚Œå€¤ãŒã‚ã‹ã‚‹ã‘ã©å…ˆã«å‰Šé™¤ã—ã¨ãï¼‰
# contours = np.delete(contours, 1678,0)
contours = np.delete(contours, -1,0)
```

OpenCVå…¬å¼ã‚µã‚¤ãƒˆâ‡¨ [Moment](http://labs.eecs.tottori-u.ac.jp/sd/Member/oyamada/OpenCV/html/py_tutorials/py_imgproc/py_contours/py_contour_features/py_contour_features.html)

```python
from math import sqrt, pi ,acos
from tqdm import tqdm
from scipy.spatial import distance

# ç²’å­ã®å½¢çŠ¶ã‚’æ¸¬å®š
Areas = []
Perimeters = []
Circularities = []
Eq_diameters = []
Longest = []
Shortest = []
Distance_Gomez = []
degrees = np.arange(0, 2*pi, 2*pi/36) # degrees: 10åº¦ãšã¤ã§å›žã™ (2pi/36 = 10deg)

for cnt in tqdm(contours):
    # é¢ç© (px*px)
    area = cv2.contourArea(cnt)
    Areas.append(area)
    
    # å‘¨å›²é•·(perimeter)
    arc = cv2.arcLength(cnt, True)
    Perimeters.append(arc)
    # å††å½¢åº¦(circularity)
    circularity = 4 * np.pi * area/(arc*arc)
    Circularities.append(circularity)
    
    # ç­‰ä¾¡ç›´å¾„(px)
    eq_diameter = np.sqrt(4 * area/np.pi)
    Eq_diameters.append(eq_diameter)
    
    # é‡å¿ƒ(Moment)
    M = cv2.moments(cnt)
    cx = int(M['m10']/M['m00'])
    cy = int(M['m01']/M['m00'])
    
    # é‡å¿ƒã‹ã‚‰è¼ªéƒ­ã¾ã§ã®è·é›¢ (Distance_Gomez)
    for deg in degrees:
        for cnts in cnt:
            # cntã®(x,y)åº§æ¨™
            x = cnts[0,0]
            y = cnts[0,1]
            # ãƒ¦ãƒ¼ã‚¯ãƒªãƒƒãƒ‰è·é›¢
            d_eu = distance.euclidean((x,y), (cx,cy))
            # æŒ‡å®šè§’åº¦ã®æ–¹å‘ã«ã‚ã‚‹è¼ªéƒ­ã®ç‚¹ã‚’æ±‚ã‚ã‚‹èª¤å·®1åº¦ã¾ã§ ( |acos(x-cx) / d_eu| < 1 degree )
            if abs(acos((x-cx)/d_eu) - deg) < pi/360:
                break
            else:
                continue
        # scaling
        d_eu_scaled = d_eu/eq_diameter
        Distance_Gomez.append(d_eu_scaled)
    # Ellipse - shortest, longest ax
    ellipse = cv2.fitEllipse(cnt)
    Shortest.append(ellipse[1][0])
    Longest.append(ellipse[1][1])
            
        
```

```python
# plot distance (n=5)
print(len(Distance_Gomez)/len(contours))
print(f'n = {len(Distance_Gomez[:180])/36}')

# plot
plt.figure(figsize=(8,3), dpi=100)
plt.plot(np.arange(len(Distance_Gomez[:180])), Distance_Gomez[:180], c='r', lw=2)
# åž‚ç›´ç·š vline
vline = np.arange(0, 181, 36)
for i in vline:
    plt.vlines([i], 0, 0.7, 'gray', linestyle='dashed', lw=1)
```

```python
# morlet wavelet n=100 (n=3600/36), omega=0.2

sns.set_style('white')
from swan import pycwt
x = np.arange(len(Distance_Gomez[:3600]))
y = Distance_Gomez[:3600]
              
Fs = 1/0.01
omega0 = 0.2

# (1)ã€€Freqã‚’æŒ‡å®šã—ã¦cwt
freqs=np.arange(0.1,10,0.025)
r=pycwt.cwt_f(y,freqs,Fs,pycwt.Morlet(omega0))
rr=np.abs(r)


plt.rcParams['figure.figsize'] = (20, 6)
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.75, 0.7, 0.2])
ax2 = fig.add_axes([0.1, 0.1, 0.7, 0.60], sharex=ax1)
ax3 = fig.add_axes([0.83, 0.1, 0.03, 0.6])

ax1.plot(x, y, 'k')

img = ax2.imshow(np.flipud(rr), extent=[0, 3600,0.1, 20], aspect='auto', cmap='jet') 
twin_ax = ax2
twin_ax.set_yscale('log')
twin_ax.set_xlim(0, 3600)
twin_ax.set_ylim(0.1, 20)
ax2.tick_params(which='both', labelleft=False, left=False)
twin_ax.tick_params(which='both', labelleft=True, left=True, labelright=False)
fig.colorbar(img, cax=ax3)
plt.show()
```

```python
from functions.pdf import best_fit_distribution, make_pdf, find_PDF

import warnings
import numpy as np
import pandas as pd
import scipy.stats as st
import statsmodels as sm
import matplotlib


matplotlib.rcParams['figure.figsize'] = (16.0, 12.0)
matplotlib.style.use('ggplot')

# Load data from statsmodels datasets
data = pd.Series(Areas)

# Plot for comparison
plt.figure(figsize=(6,3), dpi=100)
ax = data.plot(kind='hist', bins=50, normed=True, alpha=0.5, label='data', legend=True, color='red')
ax.legend(loc='upper right')

# Save plot limits
dataYLim = ax.get_ylim()

# Find best fit distribution
best_fit_name, best_fit_params = best_fit_distribution(data, 200, ax)
best_dist = getattr(st, best_fit_name)

# Update plots
ax.set_ylim(dataYLim)
ax.set_title(u'Area')
ax.set_xlabel(u'n')
ax.set_ylabel('')

# Make PDF with best params 
pdf = make_pdf(best_dist, best_fit_params)

# Display
plt.figure(figsize=(6,3), dpi=100)
ax = pdf.plot(lw=2, label='PDF', legend=True)
data.plot(kind='hist', bins=50, normed=True, alpha=0.5, label='Data', legend=True, ax=ax)

param_names = (best_dist.shapes + ', loc, scale').split(', ') if best_dist.shapes else ['loc', 'scale']
param_str = ', '.join(['{}={:0.2f}'.format(k,v) for k,v in zip(param_names, best_fit_params)])
dist_str = '{}({})'.format(best_fit_name, param_str)

ax.set_title(u'Area\n' + dist_str)
ax.set_xlabel(u'n')
ax.set_ylabel('')
```
